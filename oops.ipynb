```python
class Employee:
    name="Harry"
    language="javascript"
    salary=50000000
     
    def getinfo(self):
        print(f"the language is {self.language} and the salary is {self.salary}")
         
    def greet(self):
        print("welcome to the company")
harry=Employee()

harry.getinfo()
harry.greet() 
```

```python
class calculator:
    def __init__(self,n):
        self.n=n
    def square(self):
        print(f"the square is {self.n**2}")
    def cube(self):
        print(f"the cube is {self.n**3}")
    def squareroot(self):
        print(f"the square root is {self.n**0.5}")



a=calculator(5)
a.square()
a.cube()
a.squareroot()
```

```python
from random import randint
class train:
    def __init__ (self,trainno):
        self.trainno=trainno
    def book(self,fro,to):
      print(f"the train number {self.trainno} is booked from {fro} to {to}")
    def getstatus(self):
      print(f"the train number {self.trainno} is on time")
    def getfare(self,fro,to):
       print(f"the fare from {fro} to {to} is {randint(100,1000)}")
t= train(12345)
t.book("delhi","mumbai")
t.getstatus()
t.getfare("delhi","mumbai")
```
