##### 1.	Add few names, one name in each row, in “name.txt file”. 
##### a.	Count no of names 
##### b.	Count all names starting with vowel 
##### c.	Find longest name 


```python
with open("name.txt", "w") as f:
    f.write("Ananya\nIshita\nOm\nUrvashi\nEsha\nRohan\n")

with open("name.txt", "r") as f:
    names = f.read().splitlines()

count_names = len(names)
vowel_count = sum(1 for name in names if name[0].lower() in "aeiou")
longest_name = max(names, key=len)

print("Total Names:", count_names)
print("Names Starting with Vowel:", vowel_count)
print("Longest Name:", longest_name)
```

##### 2.	Store integers in a file. 
##### a.	Find the max number 
##### b.	Find average of all numbers 
##### c.	Count number of numbers greater than 100 


```python
with open("numbers.txt", "w") as f:
    f.write("10 200 45 150 300 80")

with open("numbers.txt", "r") as f:
    numbers = list(map(int, f.read().split()))

print("Maximum Number:", max(numbers))
print("Average:", sum(numbers)/len(numbers))
print("Count Greater than 100:", len([n for n in numbers if n > 100]))
```

##### 3.	Assume a file city.txt with details of 5 cities in given format (cityname population(in lakhs) area(in sq KM) ): 
##### Example: 
##### Dehradun 5.78 308.20 
##### Delhi 190 1484 
##### …………… 
##### Open file city.txt and read to: 
##### a.	Display details of all cities 
##### b.	Display city names with population more than 10Lakhs 
##### c.	Display sum of areas of all cities 


```python
with open("city.txt", "w") as f:
    f.write("Dehradun 5.78 308.20\n")
    f.write("Delhi 190 1484\n")
    f.write("Mumbai 124 603\n")
    f.write("Chennai 110 426\n")
    f.write("Pune 35 331\n")

with open("city.txt", "r") as f:
    cities = f.readlines()

print("All City Details:")
for city in cities:
    print(city.strip())

total_area = 0
print("\nCities with Population > 10 Lakhs:")
for city in cities:
    name, population, area = city.split()
    if float(population) > 10:
        print(name)
    total_area += float(area)

print("\nTotal Area:", total_area)
```

##### 4.	Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.  
##### Sample input 
##### 1 0 
##### 2 $ 
##### 3 1  
##### Sample Output: 
##### Error Code: integer division or modulo by zero  
##### Error Code: invalid literal for int() with base 10: '$' 3 


```python
test_cases = [(1,0), (2,'$'), (3,1)]

for a,b in test_cases:
    try:
        print(int(a)//int(b))
    except ZeroDivisionError as e:
        print("Error Code:", e)
    except ValueError as e:
        print("Error Code:", e)
```
