##### 1.	Write a Python function to find the maximum and minimum numbers from a sequence of numbers.  (Note: Do not use built-in functions.) 

```python
def find_max_min(numbers):
    max_num = numbers[0]
    min_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
        if num < min_num:
            min_num = num
    return max_num, min_num

nums = list(map(int, input().split()))
print(find_max_min(nums))
```

##### 2.	Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.  

```python
def sum_of_cubes(n):
    s = 0
    for i in range(1, n):
        s += i**3
    return s

n = int(input())
print(sum_of_cubes(n))
```

##### 3.	Write a Python function to print 1 to n using recursion. (Note: Do not use loop) 

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n-1)
    print(n, end=" ")

n = int(input())
print_numbers(n)
```

##### 4.	Write a recursive function to print Fibonacci series upto n terms. 

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

def print_fib(n, i=0):
    if i == n:
        return
    print(fib(i), end=" ")
    print_fib(n, i+1)

n = int(input())
print_fib(n)
```

##### 5.	Write a lambda function to find volume of cone. 

```python
volume = lambda r, h: (1/3) * 3.14 * r * r * h

r = float(input())
h = float(input())
print(volume(r, h))
```

##### 6.	Write a lambda function which gives tuple of max and min from a list. 

```python
lst = [10, 6, 8, 90, 12, 56]
result = (lambda x: (max(x), min(x)))(lst)
print(result)
```

##### 7.	Write functions to explain mentioned concepts: 
##### a.	Keyword argument 
##### b.	Default argument 
##### c.	Variable length argument

##### a. Keyword Argument

```python
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=18, name="Aman")
```

##### b.	Default argument 

```python
def greet(name="Guest"):
    print("Hello", name)

greet("Riya")
greet()
```

##### c.Variable Length Argument

```python
def add(*numbers):
    total = 0
    for n in numbers:
        total += n
    print(total)

add(2, 4, 6, 8)
```
