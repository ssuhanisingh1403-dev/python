##### 1.	Scan n values in range 0-3 and print the number of times each value has occurred. 

```python
n = int(input("Enter number of values: "))
values = []

for i in range(n):
    x = int(input("Enter value (0-3): "))
    values.append(x)

for i in range(4):
    print("Count of", i, "=", values.count(i))
```

##### 2.	Create a tuple to store n numeric values and find the average of all values. 

```python
n = int(input("Enter number of values: "))
nums = []

for i in range(n):
    num = float(input("Enter number: "))
    nums.append(num)

t = tuple(nums)
average = sum(t) / len(t)

print("Tuple:", t)
print("Average:", average)
```

##### 3.	WAP to input a list of scores for N students in a list data type. Find the score of the runner-up and print the output. 
##### Sample Input 
##### N = 5 
##### Scores= 2 3 6 6 5 
##### Sample output 
##### 5 
##### Note: Given list is [2, 3, 6, 6, 5]. The maximum score is 6, second maximum is 5. Hence, we print 5 as the runner-up score.


```python

N = int(input("Enter number of students: "))

scores = list(map(int, input("Enter the scores separated by space: ").split()))

unique_scores = list(set(scores))


unique_scores.sort()


runner_up = unique_scores[-2]

print("Runner-up score is:", runner_up)
```

##### 4.	Create a dictionary of n persons where key is name and value is city.  
#####  a.	Display all names 
##### b.	Display all city names 
##### c.	Display student names and cities of all students. 
##### d.	Count the number of students in each city. 


```python
n = int(input())
d = {}

for i in range(n):
    name = input()
    city = input()
    d[name] = city

for name in d.keys():
    print(name)

for city in d.values():
    print(city)

for name, city in d.items():
    print(name, city)

count = {}
for city in d.values():
    if city in count:
        count[city] += 1
    else:
        count[city] = 1

for city, c in count.items():
    print(city, c)
```

##### 5.	Store details of n movies in a dictionary by taking input from the user. Each movie must store details like name,  year, director name, production cost, collection made (earning) & perform the following :- 
##### a.	print all movie details 
##### b.	display name of movies released before 2015 
##### c.	print movies that made a profit. 
##### d.	print movies directed by a particular director.

```python
n = int(input())
movies = {}

for i in range(n):
    name = input()
    year = int(input())
    director = input()
    cost = float(input())
    collection = float(input())
    movies[name] = {
        "year": year,
        "director": director,
        "cost": cost,
        "collection": collection
    }

for name, details in movies.items():
    print(name, details)

for name, details in movies.items():
    if details["year"] < 2015:
        print(name)

for name, details in movies.items():
    if details["collection"] > details["cost"]:
        print(name)

director_name = input()
for name, details in movies.items():
    if details["director"] == director_name:
        print(name)
```
