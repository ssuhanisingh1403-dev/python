```python
def top_k(items,k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]
```

### Question 2 (10 Marks)

```python
def count_by_category(pairs):
    d={}
    for cat,word in pairs:
        d.setdefault(cat,{})
        d[cat].setdefault(word,0)
        d[cat][word]+=1
    return d
```

```python
def missing_number(nums):
    n=len(nums)+1
    return n*(n+1)//2 - sum(nums)
    
    
```

### Question 4 (10 Marks)

```python
def add_unique(x,seen=None):
    if seen is None:
        seen=set()
        if x not in seen:
            seen.add(x)
            return list(seen)
```

```python
def moving_average(stream, w):
    result = []
    window_sum = 0
    
    for i in range(len(stream)):
        window_sum += stream[i]
        
       
        if i >= w:
            window_sum -= stream[i - w]
        
        
        current_window_size = min(i + 1, w)
        result.append(window_sum / current_window_size)
    
    return result
```
