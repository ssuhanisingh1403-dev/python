##### 1. Declare these variables (x, y and z) as integers. Assign a value of 9 to x. Assign 
##### a value of 7 to y, perform addition, multiplication, division and subtraction on 
##### these two variables and print out the result.  

```python
x = 9
y = 7

z = x + y
print("Addition:", z)

z = x * y
print("Multiplication:", z)

z = x / y
print("Division:", z)

z = x - y
print("Subtraction:", z)
```

##### 2. Write a Program where the radius is taken as input to compute the area of a 
circle.  

```python
radius = float(input("Enter the radius: "))

area = 3.14159 * radius * radius

print("Area of the circle:", area)
```

##### 4.Write a program to compute the length of the hypotenuse (c) of a right triangle 
##### using Pythagoras theorem. 

```python
a = float(input("Enter length of side a: "))
b = float(input("Enter length of side b: "))

c = (a**2 + b**2) ** 0.5

print("Length of hypotenuse:", c)
```

5.Write a program to find simple interests. 

```python
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest (%): "))
time = float(input("Enter time (years): "))

simple_interest = (principal * rate * time) / 100

print("Simple Interest:", simple_interest)
```

##### 6.Write a program to find area of triangle when length of sides are given. 

```python
a = float(input("Enter length of side a: "))
b = float(input("Enter length of side b: "))
c = float(input("Enter length of side c: "))

s = (a + b + c) / 2
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

print("Area of the triangle:", area)
```

##### 7.Write a program to convert given seconds into hours, minutes and remaining 
seconds.  

```python
total_seconds = int(input("Enter total seconds: "))

hours = total_seconds // 3600
remaining_seconds = total_seconds % 3600

minutes = remaining_seconds // 60
seconds = remaining_seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", seconds)
```

##### 8. Write a program to swap two numbers without taking additional variable.  

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

a, b = b, a

print("After swapping:")
print("First number:", a)
print("Second number:", b)
```

##### 9.Write a program to find sum of first n natural numbers

```python
n = int(input("Enter a positive integer n: "))

sum_natural = n * (n + 1) // 2

print("Sum of first", n, "natural numbers is:", sum_natural)
```

##### 10. Write a program to print truth table for bitwise operators (&, |, and ^ operators) 

```python
print("A B  A&B  A|B  A^B")
for A in [0, 1]:
    for B in [0, 1]:
        print(A, B, A & B, A | B, A ^ B)
```

##### 11. Write a program to find left shift and right shift values of a given number.  

```python
num = int(input("Enter a number: "))
shift = int(input("Enter number of positions to shift: "))

left_shift = num << shift
right_shift = num >> shift

print("Left shift:", left_shift)
print("Right shift:", right_shift)
```

##### 12 Using membership operator find whether a given number is in sequence 
##### (10,20,56,78,89)  

```python
num = int(input("Enter a number: "))
sequence = (10, 20, 56, 78, 89)

if num in sequence:
    print(num, "is in the sequence")
else:
    print(num, "is not in the sequence")
```

##### 13. Using membership operator find whether a given character is in a string

```python
char = input("Enter a character: ")
string = input("Enter a string: ")

if char in string:
    print(char, "is present in the string")
else:
    print(char, "is not present in the string")
```
