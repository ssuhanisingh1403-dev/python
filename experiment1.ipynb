##### 4. Take different data types and print values using print function.

```python
# String
name = "Alice"

# Integer
age = 30

# Float
height = 5.7

# Boolean
is_student = True

# List
fruits = ["apple", "banana", "cherry"]

# Dictionary
person = {"name": "Bob", "age": 25}

# Print all variables
print(name)
print(age)
print(height)
print(is_student)
print(fruits)
print(person)
```
