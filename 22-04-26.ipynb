##### Q1. Class Variable vs Instance Variable

```python
class student:
    school_name="upes"
    def __init__(self,name,age):
        self.name=name
        self.age=age
s1=student("satyarth",20)
s2=student("riya",21)
print("school",s1.school_name)
print("school",s2.school_name)
print("student1",s1.name,s1.age)
print("student2",s2.name,s2.age)
```

##### Q2. Instance Method and Class Method

```python
class employee:
    company_name="google"
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def show_details(self):
            print(f"name:{self.name},salary:{self.salary},company:{self.company_name}")
            @classmethod
            def change_company(cls,new_company):
                cls.company_name=new_company
emp1=employee("saity",5000)
emp2=employee("riya",6000)
employee.change_company("microsoft")
emp1.show_details()
emp2.show_details()
```

##### Q3. Static Method

```python
class MathHelper:
    @staticmethod
    def is_even(num):
        return num%2==0
print(MathHelper.is_even(4))
print(MathHelper.is_even(5))
```

##### Q4. Public and Private Data Members

```python
class BankAccount:
    def __init__(self,account_number,balance):
        self.account_number=account_number
        self.balance=balance
    def deposit(self,ammount):
        self.balance+=ammount
        print(f"deposited {ammount},new balance:{self.balance}")
        
    def withdraw(self,amount):
        if amount<=self.balance:
            self.balance-=amount
            print(f"withdrew {amount},new balance:{self.balance}")
        else:
            print("insufficient balance")
    def show_balance(self):
        print(f"account number:{self.account_number},balance:{self.balance}")
        acc=BankAccount("123456789",1000)
        acc.show_balance()

        
```
